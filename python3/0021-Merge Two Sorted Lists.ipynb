##### You are given the heads of two sorted linked lists list1 and list2.

##### Merge the two lists in a one sorted list. The list should be made by splicing together the nodes of the first two lists.

##### Return the head of the merged linked list.
<br>  

**Constraints:**

* The number of nodes in both lists is in the range [0, 50].
* -100 <= Node.val <= 100
* Both list1 and list2 are sorted in non-decreasing order.



**Example 1:**  
![](../_pictures/0021.jpg)
```
Input: list1 = [1,2,4], list2 = [1,3,4]
Output: [1,1,2,3,4,4]
```  

**Example 2:**  
```
Input: list1 = [], list2 = []
Output: []
```

**Example 3:**  
```
Input: list1 = [], list2 = [0]
Output: [0]
```

````python
# 取出 & 放回

# Definition for singly-linked list.
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
        
class Solution:
    def mergeTwoLists(self, list1: Optional[ListNode], list2: Optional[ListNode]) -> Optional[ListNode]:
        list_merged = []
        
        while list1:
            list_merged.append(list1.val)
            list1 = list1.next
        while list2:
            list_merged.append(list2.val)
            list2 = list2.next
        
        list_merged.sort()
        
        res = None
        while list_merged:
            res = ListNode(list_merged.pop(), res)
        
        return res
````

````python
class Solution:
    def mergeTwoLists(self, list1: Optional[ListNode], list2: Optional[ListNode]) -> Optional[ListNode]:
        if not list1 and not list2:
            return None
        
        res = ListNode(None)
        curr = res
        i = list1
        j = list2
        
        while i and j:
            if i.val <= j.val:
                curr.next = i
                i = i.next
            else:
                curr.next = j
                j = j.next
            curr = curr.next
        
        while i:
            curr.next = i
            curr = curr.next
            i = i.next
        while j:
            curr.next = j
            curr = curr.next
            j = j.next
        
        return res.next
````
