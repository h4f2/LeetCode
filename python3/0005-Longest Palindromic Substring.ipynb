##### Given a string s, return the longest palindromic substring in s.
<br>  

**Constraints:**  
* 1 <= s.length <= 1000
* s consist of only digits and English letters.
   

**Example 1:**  
`Input: s = "babad"`  
`Output: "bab"`  
`Explanation: "aba" is also a valid answer.`

**Example 2:**  
`Input: s = "cbbd"`  
`Output: "bb"`

```python
testcase = ('babad',
            'cbbd',
            'a',
            'bb',
            'ccc')

answer = (['bab', 'aba'],
          ['bb'],
          ['a'],
          ['bb'],
          ['ccc'])
```

```python
import time

def test(testcase, answer):
    solution = Solution()
    _result = []
    _time = 0

    for i, case in enumerate(testcase):
        start = time.perf_counter()
        output = solution.longestPalindrome(case)
        process_time = time.perf_counter() - start
        result = 'Pass' if output in answer[i] else 'Fail'

        _time += process_time
        _result.append(result)
        
        print(f'Case {i+1} : input  -> {case} ')
        print(f'\t output -> {output}{" "*15}')
        print(f'Result : {result}  Time : {process_time:.7f}\n')
    
    _result = 'Pass' if 'Fail' not in _result else 'Fail'
    print('-' * 50 + f'\nResult : {_result}  Time : {_time:.7f}')
```

```python
# 補值
class Solution:
    def longestPalindrome(self, s: str) -> str:
        _s = '-' + '-'.join(i for i in s) +'--'
        longest_s = s[0]
        
        for i in range(len(_s)):
            for j in range(len(_s)):
                if i - j >= 0 and i + j < len(_s):
                    if _s[i-j] == _s[i+j]:
                        if len(_s[i-j : i+j+1]) > len(longest_s):
                            longest_s = _s[i-j : i+j+1]
                    else:
                        break
        return (''.join(longest_s.split('-')))
    
test(testcase,answer)
```

```python
# 拆解奇偶數

class Solution:
    def longestPalindrome(self, s: str) -> str:
        res = s[0]

        for i in range(len(s)-1):
            if s[i] == s[i+1]:
                for j in range(len(s)):
                    if i - j >= 0 and i + j + 1 < len(s):
                        if s[i-j] == s[i+j+1]:
                            if len(s[i-j : i+j+2]) > len(res):
                                res = s[i-j : i+j+2]
                        else:
                            break
            for j in range(len(s)):
                if i - j >= 0 and i + j < len(s):
                    if s[i-j] == s[i+j]:
                        if len(s[i-j : i+j+1]) > len(res):
                            res = s[i-j : i+j+1]
                    else:
                        break
        return res
    
test(testcase,answer)
```

```python
# 拆解奇偶數, 比較長度

class Solution:
    def longestPalindrome(self, s: str) -> str:
        s = '-' + s
        res = s[1]

        for i in range(1,len(s)-1):
            if s[i] == s[i+1]:
                for j in range((len(res)-1)//2,len(s)):
                    if i - j >= 0 and i + j + 1 < len(s):
                        if s[i+1:i-j-1:-1] == s[i:i+j+2]:
                            res = s[i-j : i+j+2]
                        else:
                            break
            for j in range((len(res)-1)//2+1,len(s)):
                if i - j >= 1 and i + j < len(s):
                    if s[i:i-j-1:-1] == s[i:i+j+1]:
                        res = s[i-j : i+j+1]
                    else:
                        break
        return res
    
test(testcase,answer)
```

```python
# 參考

class Solution():
    def longestPalindrome(self, s):
        start, maxLen, center, lenS = 0, 1, 0, len(s)
        while center < lenS:
            if lenS - center <= maxLen / 2:
                break
                
            left, right = center, center
            
            while right < lenS - 1 and s[right] == s[right+1]:
                right += 1
                
            center = right + 1
            
            while right < lenS - 1 and left and s[left-1]==s[right+1]:
                right, left = right + 1, left - 1
                
            if right - left + 1 > maxLen:
                start, maxLen = left, right - left + 1
                
        return s[start: start + maxLen]
    
test(testcase,answer)
```
